### Q1. In the Titanic dataset, fill missing values in the embarked column with the mode of that column

```python
titanic_df = pd.read_csv('titanic.csv')
titanic_df
```

```python
titanic_df['Embarked'] = titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode())
```

### Q2. In the Titanic dataset, normalize the fare column to have values between 0 and 1.

```python
titanic_df['Fare'] = (titanic_df['Fare'] - titanic_df['Fare'].min()) / (titanic_df['Fare'].max() - titanic_df['Fare'].min())
titanic_df
```

### Q3.In the Titanic dataset, create a new feature is_minor that indicates if the passenger is under 18 years old

```python
titanic_df['is_minor'] = titanic_df['Age'] < 18
titanic_df['is_minor'] = titanic_df['is_minor'].astype(bool)
titanic_df.head(10)
```

### Q4. In the Iris dataset, replace missing values in any column with the median of that column

```python
iris_df = pd.read_csv('Iris.csv')
numeric_columns = iris_df.select_dtypes(include=['number']).columns
iris_df[numeric_columns] = iris_df[numeric_columns].fillna(iris_df[numeric_columns].median())
iris_df
```

### Q5.In the Iris dataset, use one-hot encoding to convert the species column to numeric values

```python
iris_df = pd.get_dummies(iris_df, columns=['Species'], prefix='Species')
```

### Q6. In the Iris dataset, standardize the sepal length (cm) column to have a mean of 0 and a standard deviation of 1

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
iris_df['sepal length (cm)'] = scaler.fit_transform(iris_df[['SepalLengthCm']])
```

### Q7. In the Boston Housing dataset, remove rows where the CHAS column has missing values 

```python
boston_df = pd.read_csv('boston.csv')
boston_df = boston_df.dropna(subset=['chas'])
```

### Q8. In the Boston Housing dataset, detect and remove outliers in the TAX column using the IQR method

```python
boston_df = pd.read_csv('boston.csv')

Q1 = boston_df['tax'].quantile(0.25)
Q3 = boston_df['tax'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

boston_df = boston_df[(boston_df['tax'] >= lower_bound) & (boston_df['tax'] <= upper_bound)]
boston_df
```

### Q9. In the Boston Housing dataset, replace missing values in the DIS column with the median value of that column

```python
boston_df['dis'].fillna(boston_df['dis'].median())
```

### Q10. In the Wine Quality dataset, drop duplicate rows based on all columns
