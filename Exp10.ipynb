```python
import pandas as pd 
import numpy as np
from scipy import stats
```

### Q1. In the Titanic dataset, replace missing values in the age column with the median age of passengers.

```python
titanic_df = pd.read_csv("titanic.csv")
titanic_df.head(10)
```

```python
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
titanic_df.head(10)
```

### Q2. In the Titanic dataset, detect outliers in the fare column using Z-scores and display them. 

```python
z_scores = stats.zscore(titanic_df['Fare'])  # Exclude NaN values for Z-score calculation
threshold = 3 # Z score threshold

# Filter rows where the absolute Z-score is greater than the threshold (indicating outliers)
outliers = titanic_df[np.abs(z_scores) > threshold]
print(outliers[['Fare']])
```

### Q3. Given the Iris dataset, identify and print rows where sepal length (cm) is an outlier using the IQR method

```python
iris_df = pd.read_csv('Iris.csv')
Q1 = iris_df['SepalLengthCm'].quantile(0.25)
Q3 = iris_df['SepalLengthCm'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = iris_df[(iris_df['SepalLengthCm'] < lower_bound) | (iris_df['SepalLengthCm'] > upper_bound)]
if(outliers.empty):
    print('No outliers')
else:
    print(outliers)
```

### Q4. Given the Iris dataset, calculate and print the Z-scores for petal width (cm) and identify outliers 

```python
z_scores = stats.zscore(iris_df['PetalWidthCm'])

threshold = 3

outliers = iris_df[abs(z_scores) > threshold]

print("Z-scores for Petal Width (cm):")
print(z_scores)

if outliers.empty:
    print("\nNo outliers found.")
else:
    print("\nOutliers detected:")
    print(outliers)
```

### Q5. For the Boston Housing dataset, drop rows where the RM (average number of rooms per dwelling) column has missing values 

```python
boston_df = pd.read_csv('boston.csv')

print("Missing values before dropping rows:",boston_df['rm'].isnull().sum())

df_cleaned = boston_df.dropna(subset=['rm'])

print("\nMissing values after dropping rows:",df_cleaned['rm'].isnull().sum())

print("\nCleaned DataFrame:")
print(df_cleaned.head())
```

### Q6. For the Boston Housing dataset, use the IQR method to detect outliers in the AGE column and list them

```python
Q1 = boston_df['age'].quantile(0.25)
Q3 = boston_df['age'].quantile(0.75)


IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = boston_df[(boston_df['age'] < lower_bound) | (boston_df['age'] > upper_bound)]

if(outliers.empty):
    print('No outliers')
else:
    print(outliers)
```

### Q7. In the Wine Quality dataset, fill missing values in the total sulfur dioxide column with the mean of that column.

```python
wine_df = pd.read_csv('winequality-red.csv')

print("Missing values before filling:")
print(wine_df['total sulfur dioxide'].isnull().sum())

mean_sulfur_dioxide = wine_df['total sulfur dioxide'].mean()

wine_df['total sulfur dioxide'].fillna(mean_sulfur_dioxide, inplace=True)

print("\nMissing values after filling:")
print(wine_df['total sulfur dioxide'].isnull().sum())

print("\nUpdated DataFrame (first few rows):")
wine_df.head()
```
